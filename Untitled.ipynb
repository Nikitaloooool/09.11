```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('covid_19_data.csv')
data
```

```python
data['Last Update'] = pd.to_datetime(data['Last Update'])
data
```

```python
data['Day'] = data['Last Update'].apply(lambda x:x.day)
data['Hour'] = data['Last Update'].apply(lambda x:x.hour)
```

```python
data[data['Day'] == 30]
data
```

```python
test_data = data[data['Day'] == 30]
test_data.head(20)
```

```python
print('Подтвержденные случаи (весь мир): ', test_data['Confirmed'].sum())
print('Смерти (весь мир): ', test_data['Deaths'].sum())
print('Выздоровления (весь мир): ', test_data['Recovered'].sum())
```

```python
#Развитие короновируса на протяжении всего времени
plt.figure(figsize=(16,6))
data.groupby('Day').sum()['Confirmed'].plot();
```

```python
#График заражения covid-19
plt.figure(figsize=(16,6))
sns.barplot(x='Day',y='Confirmed',data=data);
```

```python
#Общие сведения по всем регионам
test_data.groupby('Country/Region').sum()
test_data
```

```python
#Регионы где нет зафиксированных случаев
data[data['Confirmed']==0]
data[data['Recovered']==0]
data.head(20)
```

```python
#Короновирус в разных странах
plt.figure(figsize=(150,140))
plt.rcParams.update({'font.size': 50})
sns.barplot(x='Recovered',y='Country/Region',data=data);
```

```python
#В каждой стране
pivoted = pd.pivot_table(data, values='Confirmed', columns='Country/Region', index='Day')
plt.rcParams.update({'font.size': 8})
pivoted.plot(figsize=(16,10));
```

```python
#В каждом городе
pivoted = pd.pivot_table(data, values='Confirmed', columns='Province/State', index='Day')
plt.rcParams.update({'font.size': 8})
pivoted.plot(figsize=(20,15));
```

```python
#Анализ данных заболеваний ковид в китае по городам
pivoted = pd.pivot_table(data[data['Country/Region']=='Mainland China'] , values='Confirmed', columns='Province/State', index='Day')
plt.rcParams.update({'font.size': 8})
pivoted.plot(figsize=(20,15))
```

```python
#Анализ данных заболеваний ковид по всему китаю
pivoted = pd.pivot_table(data[data['Country/Region']=='Mainland China'] , values='Confirmed', columns='Country/Region', index='Day')
plt.rcParams.update({'font.size': 8})
pivoted.plot(figsize=(20,15))
```
